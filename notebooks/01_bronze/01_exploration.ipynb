# Importing the Neccessary Modules

```python
import pandas as pd
import importlib.util
import sys
import os
import warnings
warnings.filterwarnings('ignore')
```

# Imoprting the neccessary File Modules
Using importlib.utils to add paths dynamically to import the neccessary modules

```python
# Define paths to all required modules
module_paths = {
    'file_path_for_load_raw_data': "../../src/data/load_raw_data.py",
    'file_path_for_save': "../../src/data/save_cleaned_data.py",
    'file_path_for_clean': "../../src/data/clean_data.py"
}

# Dynamically import all modules
loaded_modules = {}
for name, path in module_paths.items():
    spec = importlib.util.spec_from_file_location(f'src.{name}', path)
    if spec is None:
        raise ImportError(f'Could not load module from {path}')
    module = importlib.util.module_from_spec(spec)
    sys.modules[f'src.{name}'] = module
    spec.loader.exec_module(module)
    loaded_modules[name] = module

# Assign readable aliases
load_module = loaded_modules['file_path_for_load_raw_data']
clean_module = loaded_modules['file_path_for_clean']
save_module = loaded_modules['file_path_for_save']

print('All modules loaded successfully!')
print("Added the module path for loading Raw data, Saving data, and Cleaning data...")
```

# Loading the Dataset
Importing the Raw Data

## Data Cleaning

Applying the CleanData class to standardize column names, remove duplicates, and handle missing values.

```python
# STEP 2: Data Cleaning
print("STEP 2: Data Cleaning")

# Initialize CleanData
cleaning_Data = clean_module.CleanData(retention, bob)

# Converting the name of the features to the snake_case method
retention_cleaned= cleaning_Data.columns_to_snake_case(retention)
bob_cleaned = cleaning_Data.columns_to_snake_case(bob)

# Dropping the duplicate values
retention_cleaned = cleaning_Data.dropping_duplicates(retention_cleaned)
bob_cleaned = cleaning_Data.dropping_duplicates(bob_cleaned)

# Filling Numerical Null values 
retention_cleaned = cleaning_Data.dropping_nulls(retention_cleaned)
bob_cleaned = cleaning_Data.dropping_nulls_subset(bob_cleaned, ['account_number'])

# Hnadling the Date columns 
bob_cleaned = cleaning_Data.handling_date_datatypes(bob_cleaned, ['agreement_start_date', 'agreement_end_date'])

print("\nCleaning complete!")

print(f"\nRetention cleaned shape: {retention_cleaned.shape}")
print(f"BoB cleaned shape: {bob_cleaned.shape}")

print(f"\nRetention columns: {retention_cleaned.columns.tolist()}")
print(f"BoB columns: {bob_cleaned.columns.tolist()}")

print("\nRetention Info:")
print(retention_cleaned.info())
print(retention_cleaned.duplicated().sum())
print(retention_cleaned.isnull().sum())

print("\nBoB Info:")
print(bob_cleaned.info())
print(bob_cleaned.duplicated().sum())
print(bob_cleaned.isnull().sum())
```
