# Bronze Layer: Data Exploration and Initial Cleaning

This notebook handles the initial loading, exploration, and basic cleaning of the raw datasets (Retention.csv and bob.csv). The cleaned datasets are saved to the bronze layer for further processing.

```python
import pandas as pd
import numpy as np
from datetime import datetime
import os
```

```python
def to_snake_case(name: str) -> str:
    name = name.strip()
    name = name.replace(' ', '_').replace('-', '_')
    name = ''.join(ch.lower() if ch.isalnum() or ch == '_' else '_' for ch in name)
    name = '_'.join(part for part in name.split('_') if part)
    return name


def normalize_columns(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()
    df.columns = [to_snake_case(col) for col in df.columns]
    return df


def parse_dates(df: pd.DataFrame, columns: list) -> pd.DataFrame:
    df = df.copy()
    for col in columns:
        if col in df.columns:
            df[col] = pd.to_datetime(df[col], errors='coerce', dayfirst=True)
    return df
```

```python
# Load Retention.csv with proper date parsing
retention_df = pd.read_csv('../../data/01_raw/Retention.csv', encoding='utf-8', encoding_errors='ignore')
retention_df = normalize_columns(retention_df)

# Parse dates in Retention.csv
date_cols_retention = ['customer_since', 'agreement_end_date', 'case_creation_date', 'resolved_date', 'registered_date', 'expected_pull_date']
retention_df = parse_dates(retention_df, date_cols_retention)

print(f"Retention.csv shape: {retention_df.shape}")
print("Retention.csv columns:")
print(retention_df.columns.tolist())
print("\nRetention.csv info:")
retention_df.info()
```

```python
# Load bob.csv with proper date parsing
bob_df = pd.read_csv('../../data/01_raw/BoB.csv', encoding='utf-8', encoding_errors='ignore')
bob_df = normalize_columns(bob_df)

date_cols_bob = ['agreement_start_date', 'agreement_end_date']
bob_df = parse_dates(bob_df, date_cols_bob)

print(f"\nbob.csv shape: {bob_df.shape}")
print("bob.csv columns:")
print(bob_df.columns.tolist())
print("\nbob.csv info:")
bob_df.info()
```

```python
# Save cleaned datasets to bronze layer
os.makedirs('../../data/02_intermediate', exist_ok=True)
retention_df.to_csv('../../data/02_intermediate/cleaned_retention.csv', index=False)
bob_df.to_csv('../../data/02_intermediate/cleaned_bob.csv', index=False)

print("\nCleaned datasets saved to ../../data/02_intermediate/")
```
