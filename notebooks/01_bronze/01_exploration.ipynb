# Bronze Layer: Data Exploration and Initial Cleaning

This notebook handles the initial loading, exploration, and basic cleaning of the raw datasets (Retention.csv and bob.csv). The cleaned datasets are saved to the bronze layer for further processing.

```python
import sys
from pathlib import Path
BASE_DIR = Path().resolve().parents[1]
sys.path.append(str(BASE_DIR))
print("Project Root:", BASE_DIR)
```

```python
import pandas as pd
import numpy as np
from datetime import datetime
from src.cleaning.basic_cleaning import basic_cleaning
```

```python
RAW_DATA = BASE_DIR / "data" / "01_raw"

bob_path = RAW_DATA / "BoB.csv"
retention_path = RAW_DATA / "Retention.csv"

bob = pd.read_csv(bob_path)
retention = pd.read_csv(retention_path, encoding="latin1")

print("BoB shape:", bob.shape)
print("Retention shape:", retention.shape)
```

```python
print("\nBoB columns:\n", bob.columns)
print("\nRetention columns:\n", retention.columns)

print("BoB: ",bob.info())
print("Retention: ",retention.info())

print(bob.head())
print(retention.head())
```

```python
bob = basic_cleaning(bob)
retention = basic_cleaning(retention)
```

```python
INTERIM_DATA = BASE_DIR / "data" / "02_intermediate"

bob.to_parquet(INTERIM_DATA / "bob_clean.parquet", index=False)
retention.to_parquet(INTERIM_DATA / "retention_clean.parquet", index=False)
```
