# Importing the Neccessary Modules

```python
import pandas as pd
import importlib.util
import sys
import os
import warnings
warnings.filterwarnings('ignore')
```

# Imoprting the neccessary File Modules
Using importlib.utils to add paths dynamically to import the neccessary modules

```python
# Define paths to all required modules
module_paths = {
    'file_path_for_load_raw_data': "../../src/data/load_raw_data.py",
    'file_path_for_save': "../../src/data/save_cleaned_data.py",
    'file_path_for_clean': "../../src/data/clean_data.py"
}

# Dynamically import all modules
loaded_modules = {}
for name, path in module_paths.items():
    spec = importlib.util.spec_from_file_location(f'src.{name}', path)
    if spec is None:
        raise ImportError(f'Could not load module from {path}')
    module = importlib.util.module_from_spec(spec)
    sys.modules[f'src.{name}'] = module
    spec.loader.exec_module(module)
    loaded_modules[name] = module

# Assign readable aliases
load_module = loaded_modules['file_path_for_load_raw_data']
clean_module = loaded_modules['file_path_for_clean']
save_module = loaded_modules['file_path_for_save']

print('All modules loaded successfully!')
print("Added the module path for loading Raw data, Saving data, and Cleaning data...")
```

# Loading the Dataset

## Choosing X and y


```python
# First, let's review available columns and select relevant features for churn analysis
print("STEP 1: Analyzing Data Structure for Feature Selection")

print("\nRetention Dataset - Key Columns for Analysis:")
retention_key_cols = ['Case ID', 'Customer Account Number', 'Current Status', 'Resolution Status', 'Branch', 'Customer Tier', 'Case Type', 'Pull Type', 'Country']
print(retention_key_cols)

print("\nBoB Dataset - Key Columns for Analysis:")
bob_key_cols = ['account_number', 'agreement_number', 'system_status', 'agreement_start_date', 'agreement_end_date', 'renewal_type', 'branch', 'product_bob', 'fee_bob', 'total_bob']
print(bob_key_cols)

# Select features for modeling
retention_X = retention[retention_key_cols].copy()
bob_X = bob[bob_key_cols].copy()

# The target variable is churn status - we'll derive this from Current Status and Resolution Status
y = retention[['Case ID', 'Customer Account Number', 'Resolution Status']].copy()

print()
print(f"Retention X shape: {retention_X.shape}")
print(f"BoB X shape: {bob_X.shape}")
print(f"Target y shape: {y.shape}")
```
