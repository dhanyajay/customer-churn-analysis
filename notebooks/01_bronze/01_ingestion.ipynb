```python
import os
import requests
```

```python
RAW_DATA_PATH = "../../data/01_raw"
```

```python
def download_file(url, file_name):
    response = requests.get(url)
    response.raise_for_status()
    
    file_path = os.path.join(RAW_DATA_PATH, file_name)
    
    with open(file_path, "wb") as f:
        f.write(response.content)
    
    print(f"{file_name} downloaded to raw layer")
```

```python
if __name__ == "__main__":
    download_file(
        "https://raw.githubusercontent.com/RaturiAR7/data_science_dataset/main/Retention.csv",
        "Retention.csv"
    )
    
    download_file(
        "https://raw.githubusercontent.com/RaturiAR7/data_science_dataset/main/BoB.xlsx",
        "BoB.xlsx"
    )
```
